### Computing mean and std for each station/year ###
And calculating the standard deviation. In this case we are not divi

```python
# a simple scale function to normalize the data-frame row-by-row
from numpy import mean, std
def scale_temps(Din):
    matrix=Din.iloc[:,3:]
    Dout=Din.loc[:,['station','year']+range(1,366)]
    Mean=mean(matrix, axis=1).values
    Dout['Mean']=Mean
    Std= std(matrix, axis=1).values
    Dout['Std']=Std
    # Decided not to normalize each year to have mean zero and std 1
    # tmp = pd.DataFrame((matrix.values - Mean[:,np.newaxis])/Std[:,newaxis],columns=range(1,366))
    # print tmp.head()
    Dout.loc[:,1:365]=matrix.values
    return Dout
Dout=scale_temps(Data)
#reorder the columns
Dout=Dout[['station','year','Mean','Std']+range(1,366)]
Dout.head()
```

```python
# demonstrating the use of the cell magic %%time, which measures the run-time of the cell.
M=Dout.loc[:,1:365].transpose()

#M=M.dropna(axis=1)
M = M.apply(pd.Series.interpolate) #linearly interpolation of data
(columns,rows)=shape(M)
Mean=mean(M, axis=1).values

print (columns,rows), shape(Mean)
C=np.zeros([columns,columns])   # Sum
N=np.zeros([columns,columns])   # Counter of non-nan entries
```

A quick check if NaN entries have been replaced.

```python
nan_per_row=sum(isnan(M),axis=0)
nan_per_row.hist(bins=100)
```

```python
%%time
for i in range(rows):
    if i % 1000==0: 
        print i
    row=M.iloc[:,i]-Mean;
    outer=np.outer(row,row)
    valid=isnan(outer)==False
    C[valid]=C[valid]+outer[valid]  # update C with the valid location in outer
    N[valid]=N[valid]+1
valid_outer=np.multiply(1-isnan(N),N>0)
cov=np.divide(C,N)
```

```python
U,D,V=np.linalg.svd(cov)
```

### <span style="color:green">Weather station density</span> ###

In order to get the density of weather stations placed around the globe, we compute a two-dimensional histogram that counts the number of stations in each bin area.

Looking at the map plot, we can see that the United States (especially the East Coast), the coasts of Australia, as well as parts of Brazil and India are relatively densely covered with weather stations.


```python
import scipy

plt.figure(figsize=(15,10),dpi=300)
m = Basemap(projection='merc',llcrnrlat=latsmin,urcrnrlat=latsmax,\
            llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='i')
m.drawcoastlines()
# draw parallels and meridians.
parallels = np.arange(-80,81,10.)
m.drawparallels(parallels,labels=[False,True,True,False])
meridians = np.arange(10.,351.,20.)
m.drawmeridians(meridians,labels=[True,False,False,True])

# compute density of weather stations via 2D histogram
xyrange = [[min(x),max(x)],[min(y),max(y)]] # data range
nx = 500
ny = 500
bins = [nx,ny] # number of bins
EarthSurface = 510e6 # surface of earth in km^2
binArea = EarthSurface/(nx*ny)
# generate histogram
hh, xedges, yedges = scipy.histogram2d(x, y, range=xyrange, bins=bins)
hh[hh < 1] = np.nan # fill the areas with no stations by NaNs

m.imshow(hh.T,extent=np.array(xyrange).flatten(),interpolation='nearest', cmap='jet', vmin=1, vmax=16)

plt.colorbar()
plt.title('No. of stations per bin area ('+str(round(binArea))+'km^2)')
plt.show()
```

### <span style="color:green">Data plotted on the world map</span> ###

The code below plots the mean temperature, standard deviation and the first five eigenvectors on a world map.

One can clearly see that the mean temperature becomes higher the closer one gets to the equator. Moreover, the standard deviation increases as one goes further away from the equator. This reflects the fact that we have seasons in both hemispheres. The reason is because the rotational axis of the Earth is tilted (for more details: https://en.wikipedia.org/wiki/Axial_tilt).

Note that the standard deviation in the Northern hemisphere is higher than in the Southern hemisphere. The reason for this asymmetry is because that ~80% of the Southern hemisphere is covered with water. The water acts like a big heat storage and moderates the temperature variations in the Southern hemisphere.

Now, if we look at the first eigenvector V0, we observe that it basically seperates the weather stations regarding their location in the two hemispheres.

### Reconstruction ###

```python
def plot_reconstructions(selection,k=50,rows=2,columns=7):
    Eig=np.matrix(U[:,:k])
    matrix=np.matrix(Dout.ix[:,1:365])-Mean
    matrix[isnan(matrix)]=0
    Prod=matrix*Eig;
    
    Recon=Eig*Prod.transpose()+Mean[:,np.newaxis]
    plt.figure(figsize=(columns*3,rows*3),dpi=300)
    j=0;
    for i in selection:
        subplot(rows,columns,j); 
        j += 1; 
        if j>=rows*columns: break
        plot(Recon[:,i])
        plot(Djoined.ix[i,1:365]);
        title(Djoined.ix[i,'station']+' / '+str(Djoined.ix[i,'year']))
        xlim([0,365])
```

Observe in the reconstructions below that the blue line fills in (extrapolation/interpolation) the places where the measurements are not available. It also reduces the fluctuations in the relative to the original line. Recall the we are using the k top eigenvectors which explain about 88% of the variance.
