## Part 1 - Station Statistics

In the following, we want to get a sense of the underlying statistics of the weather data set. This will help us to decide which stations or measurement series to consider for the principal component analysis (PCA).

For this purpose, we have created a table that contains information about how many days of measurement each station has in a specific year for a given measurement type. The table has been generated by a MapReduce job and is saved in the file ``station_statistics.pkl``.

```python
import zlib,base64
station_stats={}
f=open('station_statistics.pkl','r')
l=f.readlines()
f.close()
for line in l:
    line = line.strip()
    station = line.split('\t')[0]
    data = line.split('\t')[1]
    station_stats[station[1:-1]]=pickle.loads(zlib.decompress(base64.b64decode(data[1:-1])))
```

```python
statstable = []
for key in station_stats.keys():
    for el in station_stats[key].items():
        statstable.append(flatten((key, el)))
```

```python
dfStats = pd.DataFrame(statstable)

dfStats.columns = ['Station', 'Type', 'Year', 'Measurements']
dfStats.to_pickle('dfStats.pkl')
```

```python
dfStats = pd.read_pickle('dfStats.pkl')
dfStats.head()
```

Using this dataframe, we can look at the distribution of the measurement series in regards with how many measurements they contain. 

```python
def plot_histograms(mtypes,columns=2,cumulative=False):
    rows = int(ceil(len(mtypes)/columns))
    plt.figure(figsize=(columns*7,rows*3),dpi=300)
    j = 1
    for i in mtypes:
        subplot(rows,columns,j) 
        if cumulative==False:
            dfStats.ix[dfStats['Type']==i,'Measurements'].hist(bins=365,range=(0.5,365.5),log=True)
            title(i)
        else:
            hist, bins = np.histogram(dfStats.ix[dfStats['Type']==i,'Measurements'],bins=365,range=(0.5,365.5))
            center = (bins[:-1]+bins[1:])/2
            cumdist = np.cumsum(hist[::-1])[::-1]
            plt.plot(center,cumdist.astype(float)/cumdist[0])
            plt.grid(True)
            plt.ylim((0.0,1.0))
            title(i+' (Total measurement series: '+str(cumdist[0])+')')
        j += 1
```

Below, we plot the five core measurement types:

       PRCP = Precipitation (tenths of mm)
       SNOW = Snowfall (mm)
	   SNWD = Snow depth (mm)
       TMAX = Maximum temperature (tenths of degrees C)
       TMIN = Minimum temperature (tenths of degrees C)
   
as well as the following measurement types:
 
       AWND = Average daily wind speed (tenths of meters per second)
       ACMC = Average cloudiness midnight to midnight from 30-second ceilometer data (percent)
	   ACMH = Average cloudiness midnight to midnight from manual observations (percent)
       ACSC = Average cloudiness sunrise to sunset from 30-second ceilometer data (percent)
	   ACSH = Average cloudiness sunrise to sunset from manual observations (percent)
	   

```python
mtypes = ['TMIN','TMAX','PRCP','SNOW','SNWD','AWND','ACMC','ACMH','ACSC','ACSH']
plot_histograms(mtypes)
```

```python
plot_histograms(mtypes,cumulative=True)
```

```python
def count_measurements(mthreshold=300):
    df = dfStats.ix[dfStats['Measurements']>=mthreshold,['Station','Type','Measurements']]
    df = df.groupby(['Station','Type'],sort=False,as_index=False).sum()
    stations = pickle.load(open('stations.pkl','rb'))
    df = df.join(stations,on='Station')
    return df
```

```python
dfMCount = count_measurements(365)
dfMCount.head()
```

```python
def plot_measurements(mtypes,vmin=0,vmax=20000):
    (latmin, latmax, lonmin, lonmax) = [-80,80,-180,180]
    rows = int(ceil(len(mtypes)/2))
    plt.figure(figsize=(16,5*rows),dpi=300)
    j = 1    
    for i in mtypes:
        df = dfMCount.ix[dfMCount['Type']==i,['Station','Measurements','latitude','longitude']]
        data = np.array(df.ix[:,'Measurements'])
        lons = np.array(df.ix[:,'longitude'])
        lats = np.array(df.ix[:,'latitude'])
        subplot(rows,2,j)
        #draw world map
        m = Basemap(projection='merc',llcrnrlat=latmin,urcrnrlat=latmax,\
            llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='i')
        m.drawcoastlines() 
        parallels = np.arange(latmin,latmax,20)
        m.drawparallels(parallels,labels=[False,True,True,False])
        meridians = np.arange(lonmin,lonmax,30)
        m.drawmeridians(meridians,labels=[True,False,False,True])
        x, y = m(lons,lats)
        plt.scatter(x, y, c=data, s=2, linewidth=0, vmin=vmin, vmax=vmax, cmap=plt.cm.jet)
        plt.colorbar()
        plt.title(i)
        j += 1
    plt.tight_layout()
```

```python
mtypes = ['TMIN','TMAX','PRCP','SNOW','SNWD','AWND','ACMH','ACSH']
plot_measurements(mtypes)
```

#### Number of measurements over time ####

Let us first look at the recording statistics by year. For this purpose, we read all required tables into a single dataframe.

```python
dfMeasurementsYear = pd.read_csv('TMINcounts_year',delimiter='\t',header=None,names=['year','N_TMIN'])

measurements = ['TMAXcounts_year','TMIN365counts_year','TMAX365counts_year']
names = ['N_TMAX','N_TMIN365','N_TMAX365']
for j in range(len(measurements)):
    df = pd.read_csv(measurements[j],delimiter='\t',index_col=0,header=None,names=[names[j]])
    dfMeasurementsYear = dfMeasurementsYear.join(df,on='year')
```

#### Select stations for given boundary ####

```python
from mpl_toolkits.basemap import Basemap, cm
import matplotlib.pyplot as plt

def select_stations(dataframe,bounds=[-80,80,-180,180]):
    lons=dataframe.ix[:,'longitude'].values
    lats=dataframe.ix[:,'latitude'].values 
    select=(lons>bounds[2]) * (lons<bounds[3])*(lats>bounds[0])*(lats<bounds[1])
    print 'Total stations selected: ', sum(select)
    return(lons[select],lats[select],select)
```

```python
complete_bounds = [-80,80,-180,180]
(latmin, latmax, lonmin, lonmax) = complete_bounds
(lons,lats,select)=select_stations(dfComplete,complete_bounds)
dfComplete = dfComplete[select]
```

#### Density of weather stations ####

```python
plt.figure(figsize=(15,10),dpi=300)
m = Basemap(projection='merc',llcrnrlat=latmin,urcrnrlat=latmax,\
            llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='i')
m.drawcoastlines()
# draw parallels and meridians.
parallels = np.arange(-80,81,10.)
m.drawparallels(parallels,labels=[False,True,True,False])
meridians = np.arange(10.,351.,20.)
m.drawmeridians(meridians,labels=[True,False,False,True])

# compute density of weather stations via 2D histogram
x, y = m(lons,lats)
xyrange = [[min(x),max(x)],[min(y),max(y)]] # data range
nx = 500 # number of bins along x
ny = 500 # number of bins along y
EarthSurface = 510e6 # surface of earth in km^2
binArea = EarthSurface/(nx*ny)
# generate histogram
hh, xedges, yedges = np.histogram2d(x, y, range=xyrange, bins=[nx,ny])
hh[hh < 1] = np.nan # fill the areas with no stations by NaNs

m.imshow(hh.T,extent=np.array(xyrange).flatten(),interpolation='nearest', cmap='jet', vmin=1, vmax=16)

plt.colorbar()
plt.title('No. of stations per bin area ('+str(round(binArea))+'km^2)')
plt.show()
```

#### Density of measurements ####

```python
def plot_measurements(names,xbins=250,ybins=250,vmin=0,vmax=20000,log=False):
    Npar = len(names)
    data = np.zeros((Npar,len(lons)))
    for i in range(Npar):
        data[i,:] = np.reshape(dfComplete.ix[:,names[i]].values[select],[1,len(lons)])

    fig=plt.figure(figsize=(16,10),dpi=300)
    for i in range(Npar):
        subplot(2,2,i)
        #draw world map
        m = Basemap(projection='merc',llcrnrlat=latmin,urcrnrlat=latmax,\
            llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='i')
        m.drawcoastlines() 
        parallels = np.arange(latmin,latmax,10) ###
        m.drawparallels(parallels,labels=[False,True,True,False])
        meridians = np.arange(lonmin,lonmax,20) ###
        m.drawmeridians(meridians,labels=[True,False,False,True])
        x, y = m(lons,lats)

        xyrange = [[min(x),max(x)],[min(y),max(y)]] # data range
        nx = xbins # number of bins along x
        ny = ybins # number of bins along y
        # generate histogram
        hh, xedges, yedges = np.histogram2d(x, y, range=xyrange, bins=[nx,ny],weights=data[i,:])
        hh[hh < 1] = np.nan # fill the areas with no stations by NaNs
        if log==True:
            hh = np.log10(hh)
        #m.imshow(hh.T,extent=np.array(xyrange).flatten(),interpolation='nearest', cmap='jet', vmin=vmin, vmax=vmax)
        #plot data points
        plt.scatter(x, y, c=data[i,:], s=2, linewidth=0, vmin=vmin, vmax=vmax, cmap=plt.cm.jet)
        plt.colorbar()
        plt.title(names[i])
    plt.tight_layout()
```

Plotting the measurements density in a log scale, we find that most of the measurement data is concentrated in North America. The density in the other continents is at least one order of magnitude lower.

```python
names = ['N_TMIN','N_TMAX','N_TMIN365','N_TMAX365']
plot_measurements(names,250,250,1,6,log=True)
```
