TASK: Try to find number of TMAX measurements for each station in ``ALL.head.csv`` via MapReduce by running mrjob locally. DONE!

The five core element types are:

       PRCP = Precipitation (tenths of mm)
       SNOW = Snowfall (mm)
	   SNWD = Snow depth (mm)
       TMAX = Maximum temperature (tenths of degrees C)
       TMIN = Minimum temperature (tenths of degrees C)

```python
%%writefile mr_weather_count_measurements.py
"""
count the number of measurements of each type
"""
from mrjob.job import MRJob

elementType = 'TMAX'

class MRWeatherCountMeasurements(MRJob):

    def mapper(self, _, line):
        self.increment_counter('MrJob Counters','mapper',1)
        elements=line.split(',')
        measurements_per_line = 0
        if elements[1]==elementType:    
            for j in range(365):
                if isinstance(elements[j+3], basestring):
                    measurements_per_line += 1        
        if elements[0]!='station':
            yield(elements[0],measurements_per_line)
            
    def combiner(self, word, counts):
        self.increment_counter('Mrjob Counters','combiner',1)
        yield (word, sum(counts))

    def reducer(self, word, counts):
        self.increment_counter('MrJob Counters','reducer',1)
        yield (word, sum(counts))

if __name__ == '__main__':
    MRWeatherCountMeasurements.run()
```

```python
from mpl_toolkits.basemap import Basemap, cm
import matplotlib.pyplot as plt

def select_stations(dataframe,bounds=[-80,80,-170,170]):
    lons=dataframe.ix[:,'longitude'].values
    lats=dataframe.ix[:,'latitude'].values
    station_names=dfComplete.ix[:,'station'].values
    (latmin, latmax, lonmin, lonmax)=bounds  
    select=(lons>lonmin) * (lons<lonmax)*(lats>latmin)*(lats<latmax)
    station_names=station_names[select]
    print 'Total stations selected: ', sum(select)
    return(lons[select],lats[select],select)
```

```python
(lons,lats,select)=select_stations(dfComplete)
```

```python
plt.figure(figsize=(15,10),dpi=300)
m = Basemap(projection='merc',llcrnrlat=latmin,urcrnrlat=latmax,\
            llcrnrlon=lonmin,urcrnrlon=lonmax,lat_ts=20,resolution='i')
m.drawcoastlines()
# draw parallels and meridians.
parallels = np.arange(-80,81,10.)
m.drawparallels(parallels,labels=[False,True,True,False])
meridians = np.arange(10.,351.,20.)
m.drawmeridians(meridians,labels=[True,False,False,True])

# compute density of weather stations via 2D histogram
x, y = m(lons,lats)
xyrange = [[min(x),max(x)],[min(y),max(y)]] # data range
nx = 500 # number of bins along x
ny = 500 # number of bins along y
EarthSurface = 510e6 # surface of earth in km^2
binArea = EarthSurface/(nx*ny)
# generate histogram
hh, xedges, yedges = np.histogram2d(x, y, range=xyrange, bins=[nx,ny])
hh[hh < 1] = np.nan # fill the areas with no stations by NaNs

m.imshow(hh.T,extent=np.array(xyrange).flatten(),interpolation='nearest', cmap='jet', vmin=1, vmax=16)

plt.colorbar()
plt.title('No. of stations per bin area ('+str(round(binArea))+'km^2)')
plt.show()
```
