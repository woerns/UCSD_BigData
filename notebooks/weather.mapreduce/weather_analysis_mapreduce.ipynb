The five core element types are:

       PRCP = Precipitation (tenths of mm)
       SNOW = Snowfall (mm)
	   SNWD = Snow depth (mm)
       TMAX = Maximum temperature (tenths of degrees C)
       TMIN = Minimum temperature (tenths of degrees C)

```python
%%writefile mr_weather_count_measurements.py
"""
count the number of measurements of each type
"""
from mrjob.job import MRJob
from sys import stderr

elementType = 'TMAX'

class MRWeatherCountMeasurements(MRJob):

    def mapper(self, _, line):
        try:
            self.increment_counter('MrJob Counters','mapper',1)
            elements=line.split(',')
            measurements_per_line = 0
            if elements[1]==elementType:    
                for j in range(365):
                    if isinstance(elements[j+3], basestring):
                        measurements_per_line += 1        
            if elements[0]!='station':
                out=(elements[0],measurements_per_line)
        except Exception, e:
            stderr.write('Error in line:\n'+line)
            stderr.write(e)
            self.increment_counter('MrJob Counters','mapper-error',1)
            out=('error',1)
        finally:
            yield out
            
    def combiner(self, word, counts):
        self.increment_counter('MrJob Counters','combiner',1)
        yield (word, sum(counts))

    def reducer(self, word, counts):
        self.increment_counter('MrJob Counters','reducer',1)
        yield (word, sum(counts))

if __name__ == '__main__':
    MRWeatherCountMeasurements.run()
```

```python
from mpl_toolkits.basemap import Basemap, cm
import matplotlib.pyplot as plt

def select_stations(dataframe,bounds=[-80,80,-170,170]):
    lons=dataframe.ix[:,'longitude'].values
    lats=dataframe.ix[:,'latitude'].values
    #station_names=dfComplete.ix[:,'station'].values
    (latmin, latmax, lonmin, lonmax)=bounds  
    select=(lons>lonmin) * (lons<lonmax)*(lats>latmin)*(lats<latmax)
    #station_names=station_names[select]
    print 'Total stations selected: ', sum(select)
    return(lons[select],lats[select],select)
```

### Space partitioning of world map using k-d tree

```python
from collections import namedtuple
from pprint import pformat

class Node(namedtuple('Node', 'key left_child right_child')):
    def __repr__(self):
        return pformat(tuple(self))

def generate_kdtree(data, depth=0, k=2, threshold=100000):
    if len(data[:,0])<=1:
        return None
    
    maxlimit = sum(data[:,2])/2
    if maxlimit<threshold:
        return None
    # Select axis based on depth so that axis cycles through dimensions
    axis = depth % k
    # Sort point list and choose median as pivot element
    idx = data[:,axis].argsort()
    data = data[idx] 

    # Compute median
    maxlimit = sum(data[:,2])/2
    counts = 0
    for j in range(len(data[:,2])):
        counts = counts + data[j,2]
        if counts + data[j+1,2] >= maxlimit:
            median = j        
            break
            
    # Create node and construct subtrees
    return Node(
        key=(data[median][0],data[median][1],counts),
        left_child=generate_kdtree(data[:median], depth + 1),
        right_child=generate_kdtree(data[median + 1:], depth + 1)
    )
```
